```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Load energy dataset
csv_file_path = 'D:/Jupyter Notebook/solar_data.csv'
energy_data = pd.read_csv(csv_file_path)

# Set 'Year' column as index
energy_data.set_index('Year', inplace=True)

# Get unique countries in the dataset
countries = energy_data['country'].unique()

# Function to predict energy generation and solar consumption for a given country
def predict_energy_and_consumption(country_data):
    # Split data into training and testing sets
    train_data = country_data.loc[:2020]  # Use data up to 2020 for training
    
    # Define forecast years
    forecast_years = range(2021, 2028)

    try:
        # Build and train ARIMA model for energy generation
        energy_model = ARIMA(train_data['Solar_generation (TWh)'], order=(7, 1, 0))  # Example order, you need to tune this
        energy_fit_model = energy_model.fit()
        energy_predictions = energy_fit_model.forecast(steps=len(forecast_years))

        # Build and train ARIMA model for solar consumption
        consumption_model = ARIMA(train_data['Solar_consumption'], order=(7, 1, 0))  # Example order, you need to tune this
        consumption_fit_model = consumption_model.fit()
        consumption_predictions = consumption_fit_model.forecast(steps=len(forecast_years))

        # Create forecast index (years)
        energy_predictions.index = forecast_years
        consumption_predictions.index = forecast_years
    
    except Exception as e:
        print(f"Error encountered for country data: {e}")
        energy_predictions = pd.Series(index=forecast_years).fillna(0)  # Fill NaN with 0 on error
        consumption_predictions = pd.Series(index=forecast_years).fillna(0)  # Fill NaN with 0 on error
    
    return energy_predictions, consumption_predictions

# Predict energy generation and solar consumption for each country
predicted_data = []

for country in countries:
    country_data = energy_data[energy_data['country'] == country]
    energy_predictions, consumption_predictions = predict_energy_and_consumption(country_data)
    
    for year in range(2021, 2028):
        predicted_data.append({
            'country': country,
            'Year': year,
            'predicted_solar_generation': energy_predictions.loc[year],
            'predicted_solar_consumption': consumption_predictions.loc[year]
        })

# Create DataFrame from predicted data
predicted_df = pd.DataFrame(predicted_data)
print(predicted_df)
# Save predictions to a CSV file
predicted_df.to_csv('solar_predicted_generation_and_consumption.csv', index=False)
```
