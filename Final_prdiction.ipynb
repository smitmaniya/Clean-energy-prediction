```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

def predict_energy_and_consumption(csv_file_path, generation_column, consumption_column):
    # Load energy dataset
    energy_data = pd.read_csv(csv_file_path)

    # Set 'Year' column as index
    energy_data.set_index('Year', inplace=True)

    # Get unique countries in the dataset
    countries = energy_data['country'].unique()

    # Function to predict energy generation and consumption for a given country
    def predict(country_data, generation_column, consumption_column):
        # Split data into training and testing sets
        train_data = country_data.loc[:2020]  # Use data up to 2020 for training

        # Define forecast years
        forecast_years = range(2021, 2028)

        try:
            # Build and train ARIMA model for energy generation
            energy_model = ARIMA(train_data[generation_column], order=(7, 1, 0))  
            energy_fit_model = energy_model.fit()
            energy_predictions = energy_fit_model.forecast(steps=len(forecast_years))

            # Build and train ARIMA model for energy consumption
            consumption_model = ARIMA(train_data[consumption_column], order=(7, 1, 0))  
            consumption_fit_model = consumption_model.fit()
            consumption_predictions = consumption_fit_model.forecast(steps=len(forecast_years))

            # Create forecast index (years)
            energy_predictions.index = forecast_years
            consumption_predictions.index = forecast_years

        except Exception as e:
            print(f"Error encountered for country data: {e}")
            energy_predictions = pd.Series(index=forecast_years).fillna(0)  x
            consumption_predictions = pd.Series(index=forecast_years).fillna(0)  

        return energy_predictions, consumption_predictions

    # Predict energy generation and consumption for each country
    predicted_data = []

    for country in countries:
        country_data = energy_data[energy_data['country'] == country]
        energy_predictions, consumption_predictions = predict(country_data, generation_column, consumption_column)

        for year in range(2021, 2028):
            predicted_data.append({
                'country': country,
                'Year': year,
                f'predicted_{generation_column.lower()}_generation': energy_predictions.loc[year],
                f'predicted_{consumption_column.lower()}_consumption': consumption_predictions.loc[year]
            })

    # Create DataFrame from predicted data
    predicted_df = pd.DataFrame(predicted_data)
    print(predicted_df)
    # Save predictions to a CSV file
    predicted_df.to_csv(f'predicted_{generation_column.lower()}_energy_and_consumption.csv', index=False)

# Predictions for Solar Energy
predict_energy_and_consumption('D:/Jupyter Notebook/solar_data.csv', 'Solar_generation (TWh)', 'Solar_consumption')

# Predictions for Wind Energy
predict_energy_and_consumption('D:/Jupyter Notebook/wind_data.csv', 'Wind_generation', 'Wind_consumption')

# Predictions for Hydro Energy
predict_energy_and_consumption('D:/Jupyter Notebook/hydro_data.csv', 'Hydro_generation ', 'Hydro_consumption')
```
